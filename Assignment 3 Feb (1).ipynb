```python
# Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
# function.

n Python, a generator function is a special type of function that generates a sequence of values using the yield keyword. A generator function returns an iterator object, which can be used to iterate over the sequence of values generated by the function.

A generator function is defined using the def keyword, like a regular function, but it uses the yield keyword instead of return to return values from the function. When a generator function is called, it does not execute the function body immediately. Instead, it returns an iterator object that can be used to iterate over the sequence of values generated by the function. Each time the yield keyword is encountered in the function body, the function returns the value specified by the yield keyword and pauses its execution. When the next value is requested from the iterator, the function resumes execution from where it left off and continues until the next yield keyword is encountered.

The yield keyword is used in a generator function to generate a value and return it to the caller without terminating the function. Unlike the return statement, which terminates the function and returns a value, the yield statement allows the function to be resumed from where it left off the next time it is called.

def squares(n):
    for i in range(n):
        yield i**2

# Using the generator function
my_iterator = squares(5)
print(next(my_iterator))  
print(next(my_iterator))  
print(next(my_iterator))  
print(next(my_iterator))  
print(next(my_iterator))
```

```python
# Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
# first 20 prime numbers.

def primes():
    num = 2
    while True:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            yield num
        num += 1

# Using the generator function
prime_generator = primes()

# Print the first 20 prime numbers using next()
for i in range(20):
    print(next(prime_generator))
```

```python
number = input("Enter a number: ")
reverse = 0
temp = int(number)

while temp > 0:
    remainder = temp % 10
    reverse = (reverse * 10) + remainder
    temp = temp // 10

if int(number) == reverse:
    print("The given number is a Palindrome")
else:
    print("The given number is not a Palindrome")

```
